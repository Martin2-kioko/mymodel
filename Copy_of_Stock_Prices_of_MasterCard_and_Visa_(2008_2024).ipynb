# Comparative Analysis

```python
# Calculate daily returns
data['Return_M'] = data['Close_M'].pct_change()
data['Return_V'] = data['Close_V'].pct_change()

# Calculate cumulative returns
data['Cumulative_Return_M'] = (1 + data['Return_M']).cumprod()
data['Cumulative_Return_V'] = (1 + data['Return_V']).cumprod()

# Plot cumulative returns
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Cumulative_Return_M'], label='MasterCard Cumulative Return')
plt.plot(data.index, data['Cumulative_Return_V'], label='Visa Cumulative Return')
plt.title('Cumulative Returns of MasterCard and Visa')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.legend()
plt.show()
```

**checking whether the dataset is stationary using Augmented Dickey -Fuller Test**

```python
from statsmodels.tsa.stattools import adfuller

# Check for stationarity using the Augmented Dickey-Fuller test
def adf_test(series):
    result = adfuller(series.dropna())
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    for key, value in result[4].items():
        print('Critial Values:')
        print(f'   {key}, {value}')

print("ADF Test for MasterCard Close Price:")
adf_test(data['Close_M'])

print("\nADF Test for Visa Close Price:")
adf_test(data['Close_V'])
```

**LSTM model**

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_percentage_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, GRU
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
```

```python
# Feature engineering
data['MA10_V'] = data['Close_V'].rolling(window=10).mean()
data['MA20_V'] = data['Close_V'].rolling(window=20).mean()
data['Volatility_V'] = data['Close_V'].rolling(window=10).std()

# Drop NaNs after rolling
data.dropna(inplace=True)
```

```python
# --- Scaling actual prices ---
scaler_M = MinMaxScaler()
scaler_V = MinMaxScaler()

scaled_data_M = scaler_M.fit_transform(data[['Close_M']])
scaled_features_V = scaler_V.fit_transform(data[['Close_V', 'MA10_V', 'MA20_V', 'Volatility_V']])
```

```python
# --- Train/test split ---
train_len_M = int(len(scaled_data_M) * 0.8)
train_len_V = int(len(scaled_features_V) * 0.8)

train_data_M = scaled_data_M[:train_len_M]
test_data_M = scaled_data_M[train_len_M:]

train_data_V = scaled_features_V[:train_len_V]
test_data_V = scaled_features_V[train_len_V:]
```

```python
# --- Sequence creation ---
def create_sequences(data, seq_length):
    x, y = [], []
    for i in range(seq_length, len(data)):
        x.append(data[i-seq_length:i])
        y.append(data[i, 0])
    return np.array(x), np.array(y)

seq_length = 60

x_train_M, y_train_M = create_sequences(train_data_M, seq_length)
x_test_M, y_test_M = create_sequences(test_data_M, seq_length)

x_train_V, y_train_V = create_sequences(train_data_V, seq_length)
x_test_V, y_test_V = create_sequences(test_data_V, seq_length)
```

```python
# --- Validation split ---
val_len_M = int(len(x_train_M) * 0.1)
val_len_V = int(len(x_train_V) * 0.1)

x_val_M, y_val_M = x_train_M[-val_len_M:], y_train_M[-val_len_M:]
x_train_M, y_train_M = x_train_M[:-val_len_M], y_train_M[:-val_len_M]

x_val_V, y_val_V = x_train_V[-val_len_V:], y_train_V[-val_len_V:]
x_train_V, y_train_V = x_train_V[:-val_len_V], y_train_V[:-val_len_V]
```

```python
# --- Model architecture (LSTM) ---
def build_model(input_shape):
    model = Sequential()
    model.add(LSTM(64, return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.2))
    model.add(LSTM(64))
    model.add(Dropout(0.2))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

callbacks = [
    EarlyStopping(patience=10, restore_best_weights=True),
    ReduceLROnPlateau(patience=5, factor=0.5)
]

model_M = build_model((60, 1))
model_V = build_model((60, 4))
```

```python
# --- Training ---
history_M = model_M.fit(x_train_M, y_train_M, validation_data=(x_val_M, y_val_M),
                        epochs=100, batch_size=32, callbacks=callbacks)

history_V = model_V.fit(x_train_V, y_train_V, validation_data=(x_val_V, y_val_V),
                        epochs=100, batch_size=32, callbacks=callbacks)
```

```python
# --- Predictions ---
pred_M = model_M.predict(x_test_M)
pred_V = model_V.predict(x_test_V)
```

```python
# Inverse transform predictions
y_test_M_rescaled = scaler_M.inverse_transform(y_test_M.reshape(-1, 1))
pred_M_rescaled = scaler_M.inverse_transform(pred_M)

y_test_V_rescaled = scaler_V.inverse_transform(
    np.hstack((y_test_V.reshape(-1, 1), np.zeros((y_test_V.shape[0], 3))))
)[:, 0]
pred_V_rescaled = scaler_V.inverse_transform(
    np.hstack((pred_V, np.zeros((pred_V.shape[0], 3))))
)[:, 0]
```

```python
# --- Evaluation ---
def evaluate_model(y_true, y_pred, name="Model"):
    print(f"{name} RMSE:", np.sqrt(mean_squared_error(y_true, y_pred)))
    print(f"{name} R^2:", r2_score(y_true, y_pred))
    print(f"{name} MAPE:", mean_absolute_percentage_error(y_true, y_pred))

evaluate_model(y_test_M_rescaled, pred_M_rescaled, "Mastercard")
evaluate_model(y_test_V_rescaled, pred_V_rescaled, "Visa")
```

```python
# --- Recreate training price data ---
train_close_M_rescaled = scaler_M.inverse_transform(train_data_M)
train_close_V_rescaled = scaler_V.inverse_transform(train_data_V)[:, 0]

# --- Date index for plotting ---
train_dates_M = data.index[:train_len_M]
test_dates_M = data.index[train_len_M + seq_length:]

train_dates_V = data.index[:train_len_V]
test_dates_V = data.index[train_len_V + seq_length:]
```

```python
# --- Mastercard Plot ---
plt.figure(figsize=(14, 6))
plt.plot(train_dates_M, train_close_M_rescaled, label='Training Data', color='blue')
plt.plot(test_dates_M, y_test_M_rescaled, label='Actual Prices', color='green')
plt.plot(test_dates_M, pred_M_rescaled, label='Predicted Prices', color='red')
plt.title("Mastercard Stock Price Prediction")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# --- Visa Plot ---
plt.figure(figsize=(14, 6))
plt.plot(train_dates_V, train_close_V_rescaled, label='Training Data', color='blue')
plt.plot(test_dates_V, y_test_V_rescaled, label='Actual Prices', color='green')
plt.plot(test_dates_V, pred_V_rescaled, label='Predicted Prices', color='red')
plt.title("Visa Stock Price Prediction")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
